```python
import matplotlib.pyplot as plt
import yfinance as yf

# TYD와 TMF의 주가 데이터를 가져오기 위한 기간 설정
start_date = '2023-01-01'
end_date = '2023-12-13'

# yfinance를 사용하여 TYD와 TMF의 주가 데이터를 가져옴
tyd = yf.download('TYD', start=start_date, end=end_date)
tmf = yf.download('TMF', start=start_date, end=end_date)

# 일일 수익률 계산
tyd_daily_returns = tyd['Close'].pct_change() * 100  # TYD의 일일 수익률 (퍼센트 단위)
tmf_daily_returns = tmf['Close'].pct_change() * 100  # TMF의 일일 수익률 (퍼센트 단위)

# 하나의 화면에 두 개의 그래프를 나타내기 위한 설정
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

# 첫 번째 그래프: TYD와 TMF의 일일 수익률 비교
ax1.plot(tyd_daily_returns, label='TYD Daily Returns', color='blue')
ax1.plot(tmf_daily_returns, label='TMF Daily Returns', color='red')
ax1.set_title('TYD vs TMF Daily Returns Comparison (2023)')
ax1.set_ylabel('Daily Returns (%)')
ax1.legend()
ax1.grid(True)

# 두 번째 그래프: TMF와 TYD의 일일 수익률 차이
returns_diff = tmf_daily_returns - tyd_daily_returns
ax2.plot(returns_diff, label='TMF - TYD Returns Difference', color='green')
ax2.axhline(0, color='gray', linestyle='--')  # 0을 기준선으로 설정
ax2.set_title('TMF - TYD Daily Returns Difference (2023)')
ax2.set_xlabel('Date')
ax2.set_ylabel('Returns Difference (%)')
ax2.legend()
ax2.grid(True)

# 전체 그래프 레이아웃 조정
plt.tight_layout()
plt.show()
```
